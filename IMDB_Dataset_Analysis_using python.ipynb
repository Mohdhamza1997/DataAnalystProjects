Creating new dataset just with the required field to tackle the questions

```python
imdb_df_new = imdb_df[['id','budget','genres','original_language','title','popularity','cast','crew','revenue']] 
```

```python
imdb_df_new.head()
```

```python
imdb_df_new.describe()
```

replacing budget and revenue which are less than 1000 to the mean of budget and revenue of movies whoes revenue are greater than 1000 respectively

```python
imdb_df_new.loc[imdb_df_new['budget']<1000,'budget'] = imdb_df_new.loc[imdb_df_new['budget']>1000,'budget'].median()
```

```python
imdb_df_new.loc[imdb_df_new['revenue']<1000,'revenue'] = imdb_df_new.loc[imdb_df_new['revenue']>1000,'revenue'].median()
```

```python
imdb_df_new.describe()
```

creating new profit column

```python
imdb_df_new['profit'] = imdb_df_new['revenue'] - imdb_df_new['budget']
```

```python
imdb_df_new.head()
```

```python
imdb_df_new.info()
```

dropping all the null values present in the dataset

```python
imdb_df_new.dropna(inplace=True)
```

```python
imdb_df_new.info()
```

Creating a function to change the string into list format for genure , crew and cast

```python
def str_to_list(item):
  return eval(item)
```

apply eval for crew, genure and cast

```python
imdb_df_new['crew'] = imdb_df_new['crew'].apply(lambda x: str_to_list(x))
```

```python
imdb_df_new['genres'] = imdb_df_new['genres'].apply(lambda x: str_to_list(x))
```

```python
imdb_df_new['cast'] = imdb_df_new['cast'].apply(lambda x: str_to_list(x))
```

Calculating the ROI and creating th column 

```python
imdb_df_new['ROI'] = (imdb_df_new['profit']/imdb_df_new['budget'])*100
```

Language with Maximum ROI

creating new column for cast named 'new cast'

<h1> Question 5 </h1>
this follow the same procedure as question number 4 but here we will find director instead of actor



```python
imdb_df_new.head()
```

```python
imdb_df_new.iloc[0]['crew']
```

```python
imdb_df_new['new_director'] = imdb_df_new.crew.fillna('[]')
```

```python
imdb_df_new.head()
```

```python
def find_director(x):
  l=[]
  for i in x:
    if i['job'] == 'Director':
      l.append(i['name'])
  return l
```

```python
imdb_df_new['new_director'] = imdb_df_new.new_director.apply(lambda x: find_director(x))
```

```python
imdb_df_new.head(20)
```

```python
sorted_roi_df = imdb_df_new.sort_values(by='ROI',ascending = False)
```
